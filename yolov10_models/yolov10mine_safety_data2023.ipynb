```python
from ultralytics import YOLO
import torch
import os
from IPython import display
from PIL import Image
%matplotlib inline
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
# Load a model
model = YOLO("yolov10m.pt")  # build a new model from scratch
```

```python
model = model.to(device)
```

```python
train_results = model.train(
    data='/home/kambhamettu.s/Projects/mining/models/yolov10_models/data4.yaml',  # path to dataset YAML
    epochs=100,   # number of training epochs
    batch = 12,  
    imgsz=640,  # training image size
    device=device,  # device to run on, i.e. device=0 or device=0,1,2,3 or device=cpu
)
```

```python
results = model.val()
print(results)
```

```python
test_img = "/home/kambhamettu.s/Projects/mining/mine_safety_data2023_yolo/images/val/0000006.jpg"
```

```python
# Perform object detection on an image
results = model(test_img)
results[0].show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load your test image
test_img = test_img  # Replace with your image path
image = cv2.imread(test_img)

# Perform object detection on the image
results = model(test_img)

# Get the output image with detections
output_image = results[0].plot()  # Create a plot with detections
```
