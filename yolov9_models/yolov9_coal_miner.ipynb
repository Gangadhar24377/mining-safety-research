```python
from ultralytics import YOLO
import torch
import os
from IPython import display
from PIL import Image
%matplotlib inline
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
%pwd
```

```python
# Load a model
model = YOLO("yolov9m.pt").to(device)
```

```python
model.info()
```

```python
model.train(
    data="/home/kambhamettu.s/Projects/mining/models/yolov9_models/data.yaml",
    epochs=100,
    batch=16,
    imgsz=640,
    device=device,
    optimizer="AdamW",  # Better for convergence
    lr0=0.001,  # Reduce learning rate
    weight_decay=0.0001,  # Reduce overfitting
    augment=True,
    mosaic=1.0,
    mixup=0.1
)
```

```python
%pwd
```

```python
test_img = "/home/kambhamettu.s/Projects/mining/coal_miner_data2023_yolo/images/val/0000006.jpg"
```

```python
import cv2
import matplotlib.pyplot as plt

# Load your test image
test_img = test_img  # Replace with your image path
image = cv2.imread(test_img)

# Perform object detection on the image
results = model(test_img)

# Get the output image with detections
output_image = results[0].plot()  # Create a plot with detections
```

```python
# Resize the image for better display (optional)
output_image = cv2.resize(output_image, (800, 600))  # Resize to 800x600

# Convert the image from BGR to RGB
output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
```

```python
# Plot the image
plt.figure(figsize=(10, 8))
plt.imshow(output_image)
plt.axis('off')  # Turn off axis
plt.show()
```
